2. How does broadcasting work in NumPy?

  * Broadcasting allows NumPy to perform operations on arrays of different shapes without explicitly reshaping them. Instead of throwing an error when shapes don't match, NumPy “broadcasts” the smaller array across the larger one so the operation can proceed.

```python
#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 72, 90, 65]
}

df = pd.DataFrame(data)

# Step 2: Define a threshold
threshold = 75

# Step 3: Filter rows where 'Score' > threshold
filtered_df = df[df['Score'] > threshold]

# Print the result
print("Original DataFrame:")
print(df)

print("\nFiltered DataFrame (Score > 75):")
print(filtered_df)
```

```python
#11.Perform matrix multiplication using NumPy?
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1: Using np.dot()
result1 = np.dot(A, B)

# Method 2: Using the @ operator (Python 3.5+)
result2 = A @ B

# Print results
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix multiplication result using np.dot():")
print(result1)

print("\nMatrix multiplication result using @ operator:")
print(result2)
```
